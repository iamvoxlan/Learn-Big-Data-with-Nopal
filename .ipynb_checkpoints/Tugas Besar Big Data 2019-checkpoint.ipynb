GHG emission inventories are developed by Parties to the Convention using scientific and methodological guidance from the Intergovernmental Panel on Climate Change (IPCC), such as 2006 IPCC Guidelines for National Greenhouse Gas Inventories, Revised Guidelines for National Greenhouse Gas Inventories (1996), IPCC Good Practice Guidance and Uncertainty Management in National Greenhouse Gas Inventories (2000) and IPCC Good Practice Guidance on Land Use, Land-use Change and Forestry (2003). # Filtering Data Berdasarkan Kategori Gas dan Negara

Fungsi di bawah digunakan untuk menyeleksi data berdasarkan kategori gas dan negara

```python
def filtering_format(country, cat):
    collect = []
    q = "SELECT * FROM gas_emission WHERE country_or_area='"+country+"' AND category ='"+cat+"' ORDER BY year"
    res = pd.read_sql_query(q,engine)
    collect.append(res)
    return collect
```

# String Indexer

```python
data = pd.read_csv('greenhouse-gas.csv')
```

```python
def stringIndexer(attr):
    uni = list(data[attr].unique())
    inp = list(data[attr].to_numpy())
    out = []
    for x in inp:
        out.append(uni.index(x))
    return out
```

```python
def multiStringIndexer(attrs):
    for x in attrs:
        data[x+'_index'] = pd.Series(stringIndexer(x),index=data.index)
```

```python
data.describe()
```

```python
multiStringIndexer(['country_or_area','category'])
```

# Section 6 - Association Rule Mining

Sebelum melakukan ARM, dilakukan proses discretization terlebih dahulu pada data value yang masih bersifat kontinu. Data value emisi gas akan dibagi menjadi 4 kategori, sangat rendah atau kategori 0 yang bernilai <25%, rendah atau kategori 1 bernilai di antara 25%-50%, tinggi atau kategori 2 bernilai di antara 50%-75% dan sangat tinggi atau kategori 3 bernilai 75% ke atas.

```python
def discretization(attr):
    inp = list(data[attr].to_numpy())
    midi = data[attr].describe()
    out = []
    for x in inp:
        if(x<=float(midi.loc['25%'])):
            out.append(0)
        elif(x>float(midi.loc['25%']) and x<=float(midi.loc['50%'])):
            out.append(1)
        elif(x>float(midi.loc['50%']) and x<=float(midi.loc['75%'])):
            out.append(2)
        else:
            out.append(3)
    return out
```

```python
def multiDiscretization(attrs):
    for x in attrs:
        data[x+'_disc'] = pd.Series(discretization(x),index=data.index)
```

```python
multiDiscretization(['value'])
```

```python
data
```

```python
data.to_sql('gas_emission_disc',engine)
```

```python
def ARMformat(a,b,tb):
    collect = []
    for x in data[a].unique():
        q = "SELECT "+b+" FROM "+tb+" WHERE "+a+"='"+x+"'"
        res = pd.read_sql_query(q,engine)
        res=res[b].unique()
        resclean = []
        for i in res:
            if i!=None:
                resclean.append(i)
        resmin = resclean[0:10] 
        collect.append(list(resmin))
    return collect
```

```python
data_ARM = ARMformat('category','value_disc','gas_emission_disc')
```
